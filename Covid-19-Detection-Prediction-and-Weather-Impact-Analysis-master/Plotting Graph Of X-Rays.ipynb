```python
from keras import models
from keras import layers
from keras.applications import VGG16
from keras import optimizers
from keras.layers.core import Flatten, Dense, Dropout, Lambda
import tensorflow as tf
import pandas as pd
```

```python
VGG19_log = pd.read_csv('trainingVGG19.log', sep=',', engine='python')

VGG16_log = pd.read_csv('trainingVGG16.log', sep=',', engine='python')

Xception_log = pd.read_csv('trainingXception.log', sep=',', engine='python')

Inception_log = pd.read_csv('trainingInceptionV3.log', sep=',', engine='python')

ResNet50_log = pd.read_csv('trainingResnet50.log', sep=',', engine='python')

VGG16_log.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

NUM_EPOCHS = 20


# plotting the train accuracy graph
vg16 = np.array(VGG16_log['acc'])
vg19 = np.array(VGG19_log['acc'])
Excep = np.array(Xception_log['acc'])
Incep = np.array(Inception_log['acc'])
Res50 = np.array(ResNet50_log['acc'])
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(range(1,NUM_EPOCHS), vg16[1:], label='VGG16')
plt.plot(range(1,NUM_EPOCHS), vg19[1:], label='VGG19')
plt.plot(range(1,NUM_EPOCHS), Excep[1:], label='Xception')
plt.plot(range(1,NUM_EPOCHS), Incep[1:], label='Inception')
plt.plot(range(1,NUM_EPOCHS), Res50[1:], label='ResNet50')
plt.title('Accuracy over '  + str(NUM_EPOCHS) + ' Epochs', size=15)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.legend()
plt.grid(True)


#plotting the test val_accuracy graph
vg16val = np.array(VGG16_log['val_acc'])
vg19val = np.array(VGG19_log['val_acc'])
Excepval = np.array(Xception_log['val_acc'])
Incepval = np.array(Inception_log['val_acc'])
Res50val = np.array(ResNet50_log['val_acc'])
plt.figure(figsize=(15, 5))
plt.subplot(122)
plt.plot(range(1,NUM_EPOCHS), vg16val[1:], label='VGG16')
plt.plot(range(1,NUM_EPOCHS), vg19val[1:], label='VGG19')
plt.plot(range(1,NUM_EPOCHS), Excepval[1:], label='Xception')
plt.plot(range(1,NUM_EPOCHS), Incepval[1:], label='Inception')
plt.plot(range(1,NUM_EPOCHS), Res50val[1:], label='ResNet50')
plt.title('Validation Accuracy over '  + str(NUM_EPOCHS) + ' Epochs', size=15)
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()




```

```python
# plotting the train accuracy graph
vg16los = np.array(VGG16_log['loss'])
vg19los = np.array(VGG19_log['loss'])
Exceplos = np.array(Xception_log['loss'])
Inceplos = np.array(Inception_log['loss'])
Res50los = np.array(ResNet50_log['loss'])
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(range(1,NUM_EPOCHS), vg16los[1:], label='VGG16')
plt.plot(range(1,NUM_EPOCHS), vg19los[1:], label='VGG19')
plt.plot(range(1,NUM_EPOCHS), Exceplos[1:], label='Xception')
plt.plot(range(1,NUM_EPOCHS), Inceplos[1:], label='Inception')
plt.plot(range(1,NUM_EPOCHS), Res50los[1:], label='ResNet50')
plt.title('Loss over '  + str(NUM_EPOCHS) + ' Epochs', size=15)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
```

```python
# plotting the train accuracy graph
vg16losval = np.array(VGG16_log['val_loss'])
vg19losval = np.array(VGG19_log['val_loss'])
Exceplosval = np.array(Xception_log['val_loss'])
Inceplosval = np.array(Inception_log['val_loss'])
Res50losval = np.array(ResNet50_log['val_loss'])
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(range(1,NUM_EPOCHS), vg16losval[1:], label='VGG16')
plt.plot(range(1,NUM_EPOCHS), vg19losval[1:], label='VGG19')
plt.plot(range(1,NUM_EPOCHS), Exceplosval[1:], label='Xception')
plt.plot(range(1,NUM_EPOCHS), Inceplosval[1:], label='Inception')
plt.plot(range(1,NUM_EPOCHS), Res50losval[1:], label='ResNet50')
plt.title('Validation Loss over '  + str(NUM_EPOCHS) + ' Epochs', size=15)
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
```
